# The Normal Distribution (Gaussian)

### Variance and Spread

**Mean**, **Mode**, and **Median** are all the same.

Gaussian is unique in having a direct known relationship b/w the standard deviation and the range of data captured across the dataset. Thus, **the standard dev for a Normal Distribution can tell you how much data is contained within its bounds.** This is often called the *68-95-99.7* rule. **IN OTHER WORDS:**
* 68% of data is captured w/i one standard deviation from the mean.
* 95% of data is captured w/i two standard deviations from the mean.
* 99.7% of data is captured w/i three standard deviations from the mean.

So our ability to grab set quantities of data thanks to the amazing cheat code of ~*standard deviation*~!

### Worktime. Oh yes!

**Verify the above statements in re: 68-95-99.7 rule when the median is 0.**

```python
import scipy.stats as st
print("Verification of the 68-95-99.7 rule when the mean is 0:")
print("One standard deviation from the mean: " + str(st.norm.cdf(1) - st.norm.cdf(-1)))
print("Two standard deviations from the mean: " + str(st.norm.cdf(2) - st.norm.cdf(-2)))
print("Three standard deviations from the mean: " + str(st.norm.cdf(3) - st.norm.cdf(-3)))
```

**Math instructor has finished grading math exams. He reports that the mean was 60 out of 100 with a standard dev of 10. What is the probability that students got more than 70? We were not given any raw data.**

*Here's what the following line of code says: What's the **percentage of students who have earned a score of 70**, when the **average score is 60 and the std. dev. is 10**?*

```python
print(1 - st.norm(60,10).cdf(70))
print(round(1 - st.norm(60, 10).cdf(70), 2))
```

### Confused? SAME. Some explanation, courtesy of Kash

**CDF** is getting the % encapsulated by the value you're sending it. That means, what percentage of students scored exactly or below this value?

Let's look at:

#### st.norm(0,1).cdf(100)



If most students are +1 or -1 (because the st. dev. is 1, which we know from looking at the above function and the placement of the parameter), and if the mean is 0, and then we go 100 pts above, its the equivalent of saying like 200000000 into the cdf fxn.

**.norm()** accepts 2 parameters: the **mean** and the **std. dev.** of our entire dataset, correspondingly.

**.cdf()** accepts 1 parameter: the **y-value you are looking for**.

**.norm() = scope, .cdf() = bullet**
